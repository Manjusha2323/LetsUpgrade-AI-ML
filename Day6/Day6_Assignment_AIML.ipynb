```python
##Question 3 - Jupyter notebook for sets
##Create a set
my_set={"red","orange","yellow","green","blue","indigo","violet"}
print(my_set)

##Iterate through the set
print("\nIterating through the set")
for x in my_set:
    print(x)
    
##check if an item is present
print("\nChecking an element in the set")

check_colour="black"
if check_colour in my_set:
    print(check_colour,"is present in the set")
else:
    print(check_colour,"is not present in the set")
    
##to add one item in set using add()
print("\nAdding one item to the set")
my_set.add("white")
print(my_set)

##to adding multiple items to the set using update()
print("\nAdding multiple items to the set")
my_set.update(["black","grey","pink"])
print(my_set)


##to get the number of items in the set
print("\nGetting the number of items in the set")
print("Number of items in the set are :",len(my_set))

##to remove an item from the set using remove
print("\nRemoving an item from the set using remove")
my_set.remove("black")
print("Removed black from the set",my_set)

##to remove an item from the set using discard
print("\nRemoving an item from the set using discard")
my_set.discard("white")
print("Removed white from the set",my_set)

##to remove an item from the set using pop
print("\nRemoving an item from the set using pop. It removes the last item and returns it")
print("Removed item" ,my_set.pop())


##to empty the set
print("\nRemoving all items from the set")
my_set.clear()
print("Removed all items from the set :",my_set)

##to delete the set
print("\nDeleting the set")
del my_set

##Operation on 2 sets
print("\nOperations on 2 sets")
set1={"onion","potato","tomato","brinjal"}
set2={"brinjal","beans","mushrooms","onion"}
print("Set1 is",set1)
print("Set2 is",set2)


## Combining items using union
print("\nCombining 2 sets using union, duplicates discarded")
set3=set1.union(set2)
print("Combined set is",set3)

## Updating set1 with set2
print("\nUpdating set1 with set2, duplicates discarded")
set1.update(set2)
print("Updated Set1 is",set1)
print("Set2 is",set2)

##Difference between 2 sets

print("\nDifference between Set1 and Set2 is",set1.difference(set2))

##Intersection between 2 sets

print("\nIntersection between Set1 and Set2 is",set1.intersection(set2))

##Symmetric difference between 2 sets
print("\nSymmetric Difference between Set1 and Set2 is",set1.symmetric_difference(set2))

##check if set2 is subset of set1
if set2.issubset(set1):
    print("\nSet1 is subset of Set2")
else:
    print("\nSet1 is not a subset of Set2")
    
##check if there are intersecting elements between set1 and set2
if set1.isdisjoint(set2):
    print("\nThere are no common items between Set1 and Set2")
else:
    print("\nThere are  common items between Set1 and Set2")
    
##check is a set is a superset
if set1.issuperset(set2):
    print("\nSet1 is superset of Set2")
else:
    print("\nSet1 is not a superset of Set2")
    
##update Set1 with difference between the 2.
set1.difference_update(set2)
print("\nAfter Difference Update")
print("Set1 is now",set1)
print("Set2 is now",set2)

##update Set1 with intersection between the 2
set1.intersection_update(set2)
print("\nAfter Intersection Update")
print("Set1 is now",set1)
print("Set2 is now",set2)


##update Set1 with symmetric difference between the 2
set1.symmetric_difference_update(set2)
print("\nAfter symmetric difference Update")
print("Set1 is now",set1)
print("Set2 is now",set2)

##Copying a set
set3=set1.copy()
print("\nSet3 is copied from Set1 ",set3)

##Creating a set using set constructor
set4=set(("Python","is","awesome"))
print("\nSet4 is created using set constructor",set4)
```

```python
#Question 4 identify missing number for a space separated sequence of numbers
import functools
from functools import reduce
input_num=input("Enter the range of numbers :")
int_list=sorted(list(map(int,input_num.strip().split(" "))))
new_list=list(range(min(int_list),max(int_list)+1))
print("The missing number is : ",reduce(lambda a,b:a+b,new_list)-reduce(lambda a,b:a+b,int_list))

```
