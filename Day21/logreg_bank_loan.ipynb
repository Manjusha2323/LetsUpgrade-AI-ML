```python
import pandas as pd
import statsmodels.api as sm
```

```python
bank_loan_data=pd.read_excel('C:/Users/Manzoo/Learning/Day21/Assignments/dataset/Bank_Personal_Loan_Modelling.xlsx',sheet_name=1)
bank_loan_data.head()
bank_loan_data.isnull().sum()
bank_loan_data.shape
bank_loan_data.drop_duplicates()
bank_loan_data.shape
```

```python
bank_loan_data.corr()
```

```python
Y=bank_loan_data['Personal Loan']
X=bank_loan_data[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','CreditCard']]
```

```python
X1=sm.add_constant(X)
```

```python
Logistic=sm.Logit(Y,X1)
```

```python
result=Logistic.fit()
```
