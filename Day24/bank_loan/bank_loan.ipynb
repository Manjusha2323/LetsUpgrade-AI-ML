```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
bank_loan_data=pd.read_excel("C:/Users/Manzoo/Learning/Day24/Assignment/Dataset/Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)

bank_loan_data.isnull().sum()
bank_loan_data.info()
```

```python
bank_loan_data.corr()
```

```python
rf_bank_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
bank_loan_data.columns
```

```python
features=['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard']
```

```python
rf_bank_model.fit(X=bank_loan_data[features],y=bank_loan_data['Personal Loan'])
```

```python
print(rf_bank_model.oob_score_)
```

```python
for itr_features,imp in zip(features,rf_bank_model.feature_importances_):
    print(itr_features,imp)
```

```python
predictors=pd.DataFrame([bank_loan_data['Income'],bank_loan_data['CCAvg'],bank_loan_data['Education'],bank_loan_data['Family']]).T
print(predictors)
```

```python
tree_model=tree.DecisionTreeClassifier(max_depth=8)
tree_model.fit(X=predictors,y=bank_loan_data['Personal Loan'])
with open('Dtree_bankloan.dot','w') as f:
    f=tree.export_graphviz(tree_model,feature_names=['Income','CCAvg','Eductaion','Family'],out_file=f)
```

```python
tree_model.score(X=predictors,y=bank_loan_data['Personal Loan'])
```
