```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
attr_data=pd.read_csv("C:/Users/Manzoo/Learning/Day24/Assignment/Dataset/general_data.csv")
attr_data.isnull().sum()
attr_data=attr_data.fillna(attr_data.mean())
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
attr_data['Gender']=label_encoder.fit_transform(attr_data['Gender'])
attr_data['BusinessTravel']=label_encoder.fit_transform(attr_data['BusinessTravel'])
attr_data['Department']=label_encoder.fit_transform(attr_data['Department'])
attr_data['JobRole']=label_encoder.fit_transform(attr_data['JobRole'])
attr_data['MaritalStatus']=label_encoder.fit_transform(attr_data['MaritalStatus'])
attr_data['EducationField']=label_encoder.fit_transform(attr_data['EducationField'])
attr_data['Attrition']=label_encoder.fit_transform(attr_data['Attrition'])
```

```python
attr_data.corr()
```

```python
rf_attr_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
attr_data.columns
```

```python
##features=['Age','BusinessTravel','Department','DistanceFromHome','Gender','Education','EducationField','Gender','JobLevel','JobRole','MaritalStatus','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','TotalWorkingYears','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']
##features=['Age','MaritalStatus','MonthlyIncome','TotalWorkingYears','YearsAtCompany','YearsWithCurrManager','JobRole']
features=['Age','MonthlyIncome','TotalWorkingYears','YearsAtCompany']
```

```python
rf_attr_model.fit(X=attr_data[features],y=attr_data['Attrition'])
```

```python
print(rf_attr_model.oob_score_)
```

```python
for features_itr,imp in zip(features,rf_attr_model.feature_importances_):
    print(features_itr,imp)
```

```python
predictors=pd.DataFrame([attr_data['Age'],attr_data['MonthlyIncome'],attr_data['TotalWorkingYears'],attr_data['YearsAtCompany'],attr_data['JobRole']]).T
##print(features)
##predictors=pd.DataFrame(attr_data[features])
print(predictors)
```

```python
tree_model=tree.DecisionTreeClassifier(max_depth=10)
tree_model.fit(X=predictors,y=attr_data['Attrition'])
with open('Dtree_attrition.dot','w') as f:
    f=tree.export_graphviz(tree_model,feature_names=['Age','MonthlyIncome','TotalWorkingYears','YearsAtCompany','JobRole'],out_file=f)
```

```python
tree_model.score(X=predictors,y=attr_data['Attrition'])
```
